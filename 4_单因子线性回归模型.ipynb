```python
import pandas as pd
data = pd.read_csv('4_1.csv')
print(data.head())
```

```python
x = data.loc[:,'x']
y = data.loc[:,'y']

print(x, y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3, 3))
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, color = 'red')
plt.title('Sigle Linear Regression')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

print(x, y) 
```

```python
print(x.shape, y.shape) #13行的一维数组
model.fit([x], [y]) #fit函数的参数必须是二维数组，也可以使用numpy转换
```

```python
# 使用numpy把一维转换成二维
import numpy as np
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
print(x.shape, y.shape)

model.fit(x, y) #现在可以正常运行了
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
print(y) # 与上面的是一样的， 说明预测没问题
```

```python
y_predict = model.predict([[5.6]]) # 注意这里的5.6是一个二维数组
print(y_predict) # 预测5.6的y值
```

#### 模型评估

```python
# a/b打印
a = model.coef_
b = model.intercept_
print(f'斜率a = {a}, 截距b = {b}')
```

```python
# MSE要尽可能小，R2越接近1越好
from sklearn.metrics import mean_squared_error, r2_score
MSE = mean_squared_error(y, y_pred) # y_pred是二维数组
R2 = r2_score(y, y_pred)

print(f'MSE = {MSE}, R2 = {R2}')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3, 3))
plt.xlabel('y')
plt.ylabel('y_pred')
plt.scatter(y, y_pred, color = 'blue')
```

> 上面可以看出斜率为1，即预测效果很好
